```python
import random
from collections import Counter
import matplotlib.pyplot as plt
from tqdm import tqdm  # Importar tqdm para la barra de carga

# Generar 1,000,000 de números aleatorios entre 1 y 20
numeros = []
for _ in tqdm(range(100_000_000), desc="Generando números...", ncols=100):  # Barra de carga
    numeros.append(random.randint(1, 20))

# Contar las frecuencias de cada número
frecuencias = Counter(numeros)

# Mostrar las frecuencias
for numero, frecuencia in sorted(frecuencias.items()):
    print(f"El número {numero} salió {frecuencia} veces.")

# Crear la gráfica
numeros_unicos = list(frecuencias.keys())
frecuencias_values = list(frecuencias.values())

plt.bar(numeros_unicos, frecuencias_values, color='skyblue')

# Personalizar la gráfica
plt.title('Frecuencia de números generados (1-20)')
plt.xlabel('Número')
plt.ylabel('Frecuencia')
plt.xticks(range(1, 21))  # Mostrar números del 1 al 20 en el eje X
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar la gráfica
plt.show()

```

```python
import numpy as np
# Calcular la desviación estándar
desviacion_estandar = np.std(numeros)
print(f"\nLa desviación estándar de los números generados es: {desviacion_estandar}")
```
